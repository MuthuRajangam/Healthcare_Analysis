7. Find the total posted in month of September?


1. Find the Distinct values of procudre code?




SELECT DISTINCT Procedure_Code, COUNT(*) AS Procedure_Code_count
FROM Healthcare
GROUP BY Procedure_Code;

2. Find the Distinct values of Insurance type using group by?

select distinct Insurance_Type, count(*)as Insurance_Type_Count from Healthcare group by Insurance_Type

3. Find the Distinct values of Claim status using group by?

select distinct Claim_Status, count(*)as Claim_Status_count from Healthcare group by Claim_Status

4. Find the Distinct values of Outcome count using group by?

select distinct Outcome, count(*)as Outcome_count from Healthcare group by Outcome

5. Find the Distinct values of AR_status using group by?

select distinct AR_Status, count(*)as AR_Status_count from Healthcare group by AR_Status

6. Find the total posted in month of August?

SELECT sum(Paid_Amount)as Aug_month_paid
FROM Healthcare
WHERE Date_of_Service between '2024-08-01' and '2024-08-30' ;

7. Find the total posted in month of September?

SELECT sum(Paid_Amount)as sep_month_paid
FROM Healthcare
WHERE Date_of_Service between '2024-09-01' and '2024-09-30' ;
